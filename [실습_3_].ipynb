```python
# 평균 제곱 오차 Mean Square Error(MSE)
# [4강] 선형회귀.pdf p.29
def mean_square_error(y_hat, y):
  ########### Fill your code ############
  return np.sum((y_hat - y)**2)
  #######################################
```

```python
# [2강] 선형회귀.pdf p.23
x = np.array([-2, -1, 0, 1, 2])
y = 2*(x**3) + 7
X = np.c_[np.ones(5), x, x**2, x**3]
Y = np.copy(y)
print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
```

```python
########### Fill your code ############
# 파랑색으로 (x, y)점 찍기
plt.plot(x,y, 'b.')
#######################################
plt.xlim(-3., 3)
plt.ylim(-10, 25)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# [2강] 선형회귀.pdf p.30
# 핸즈온 머신러닝 p.160
eta = 0.01
n_iterations = 1000
m = 5

theta = np.random.randn(4)
print("Before")
print(theta)

# Gradient descent Rule
for iteration in range(n_iterations):
  ########### Fill your code ############
  gradients = 2/m * X.T @ (X@theta-Y)
  theta = theta- eta*gradients
  #######################################

print("\nAfter")
print(theta)
```

```python
y_hat = X.dot(theta)
print(y_hat)
```

```python
plt.plot(x, y_hat, 'r-')
plt.plot(x, y, 'b.')
plt.axis([-3, 3, -10, 25])
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x_ = np.linspace(-3, 3, 100)
########### Fill your code ############
y_hat = theta[0] + theta[1] *x_  + theta[2] *x_**2  + theta[3] * x_**3
#######################################
plt.plot(x_, y_hat, 'r-')
plt.plot(x, y, 'b.')
plt.axis([-3, 3, -10, 25])
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# 3. Linear Regression Example
핸즈온 머신러닝 p.160~

```python
# 핸즈온 머신러닝 p.160
np.random.seed(160)
x = 2 * np.random.rand(1, 1)          # x = [0, 2] 사이의 random 값
y = 4 + 3 * x + np.random.randn(1, 1) # y = 3x + 4 + noise
```

```python
plt.plot(x, y, 'y.')
plt.xlim(0., 2.0)
plt.ylim(0, 14)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

###3.1 Linear Regression(Analytical)

```python
X = np.c_[np.ones((1, 1)), x]
Y = np.copy(y)
```

image.png

###3.2 Linear Regression(Numerical)

```python
X = np.c_[np.ones((10, 1)), x]
Y = np.copy(y)
```

image.png

```python
# [2강] 선형회귀.pdf p.30
# 핸즈온 머신러닝 p.160
eta = 0.1
n_iterations = 1000
m = 100

theta = np.random.randn(2, 1)
print("Before")
print(theta)

# Gradient descent Rule
for iteration in range(n_iterations):
  ########### Fill your code ############
  gradients = 2/m * X.T @ (X@theta-Y)
  theta = theta- eta*gradients
  #######################################

print("\nAfter")
print(theta)
```

```python
x_ = np.linspace(0, 2, 10)
y_hat = theta[0] + theta[1] * x_
plt.plot(x_, y_hat, 'r-')
plt.plot(x, y, 'y.')
plt.axis([-0, 2, 0, 14])
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 3.3 Gradient Descent Rule Update Visualization

```python
# 핸즈온 머신러닝 p.160
eta = 0.02
n_iterations = 15
m = 100

np.random.seed(160)
theta = np.random.randn(2, 1)
for iteration in range(n_iterations):
    gradients = 2/m * X.T.dot(X.dot(theta) - Y)
    new_theta = theta - eta * gradients


    X_new = np.array([[0], [2]])
    X_new_b = np.c_[np.ones((2, 1)), X_new]
    y_predict = X_new_b.dot(theta)
    y_predict

    if iteration == 0:
      plt.plot(X_new, y_predict, 'g-')
    elif iteration == n_iterations-1 :
      plt.plot(X_new, y_predict, 'b-')
    else :
      plt.plot(X_new, y_predict, 'r-')

    theta = new_theta

plt.plot(x, y, 'y.')
plt.axis([0, 2, -1, 12])
plt.grid()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

### 3.4 Linear Regression With scikit-learn
핸즈온 머신러닝 p.162

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
np.random.seed(160)
x = 2 * np.random.rand(100, 1)          # x = [0, 2] 사이의 random 값
y = 4 + 3 * x + np.random.randn(100, 1) # y = 3x + 4 + noise
```

```python
lin_reg.fit(x, y)
```

```python
lin_reg.intercept_, lin_reg.coef_ #intercept절편, coef기울기
```

```python
x_new = np.array([[0], [2]])
lin_reg.predict(x_new)
```

```python
plt.plot(x_new, lin_reg.predict(x_new), 'r-')
plt.plot(x, y, 'y.')
plt.axis([0, 2, 0, 15])
plt.grid()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
